```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Genenrate Sample Dataset
data={
    'ID': range(1,11),
    'Longitude':np.random.uniform(-1.3,-1.2,10), # Random latitudes
    'Longitude':np.random.uniform(36.7,36.9,10), # Random longitudes
    'Value':np.random.randint(10,100,10) # Random values
}
df=pd.DataFrame(data)
```

```python
# Exploratory Data Ananlysis
print("Dataset Overview:\n",df.head())
print("\nSummary Statistics:\n",df.describe())
print("\nMissing Values:\n",df.isnull().sum())
```

```python
# Box PLOT
plt.figure(figsize=(8,6))
sns.boxplot(data=df[['Value']])
plt.title("Box Plot of Vlues")
plt.show()
```

```python
# Pair Plot
sns.pairplot(df)
plt.show()
```
