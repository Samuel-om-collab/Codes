```python
# Import necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import statsmodels.api as sm
```

```python
# Load California Housing dataset
from sklearn.datasets import fetch_california_housing
california=fetch_california_housing()
df=pd.DataFrame(california.data,columns=california.feature_names)
df["PRICE"]=california.target # Target variable(median house value)
```

```python
# Display first few rows (5 always)
print("Dataset Preview.")
print(df.head())
```

```python
# Checking missing values
print("\nmissing Values:")
print(df.isnull().sum())
```

```python
# Summary stastistics
print("\nDataset Summary:")
print(df.describe())
```

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Split dataset into features (x) and target (y)
x=df.drop("PRICE",axis=1) # Independent variables
y=df["PRICE"] # Dependent variables
```

```python
# Train-test split(80% training,20% testing)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Feature scalling (Standardization)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
# Train the multiple linear regression model
model=LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
# Prediction
y_pred=model.predict(x_test_scaled)
```

```python
# Model Evaluation
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
```

```python
# AIC and BIC Clculation ( using statmodels) ARE  essential tools in multiple linear regression for selecting the best model
x_train_const=sm.add_constant(x_train_scaled) #Add intercept
ols_model=sm.OLS(y_train,x_train_const).fit()
aic=ols_model.aic
bic=ols_model.bic
```

```python
# Print performamnce metrics
print("\nill Model Evaluation Metrics:")
print(f"R-squared (R2): {r2:.4f}")
print(f"Mean Absolute Error(MAE): {mae:.4f}")
print(f"Mean Squared Error(MSE): {mse:.4f}")
print(f"Root Mean Squared Error(RMSE): {rmse:.4f}")
print(f"Bayesian Information Criterion (BIC): {bic:.4f}")
print(f"Akaike Information Criterion (AIC): {aic:.4f}")
```

```python
# Plot actual vs predicted values
plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred,alpha=0.7,color="blue")
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],linestyle="--",color="red")#Line y=x
plt.show()
```
